```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("data_cleaning_practice_dataset.xlsx");

print(df.shape)
print("Column Names : ", df.columns.to_list());

#Q1 Find missing values column-wise.
for col in df.columns.to_list():
    print(f"Missing value in {col}", df[col].isna().sum());

#Q2 Display rows containing missing values.
print(df[df.isna().any(axis=1)]);

#Q3 Replace missing Age values.
df['Age'] = df['Age'].fillna(df['Age'].median())
print(df['Age']);

#Q4 Replace missing Salary values.
df['Salary'] = df['Salary'].fillna(df['Salary'].median());

#Q5 Find duplicate rows.
print(df[df.duplicated()]);

#Q6 Remove duplicate rows.
df.drop_duplicates(inplace=True);

#Q7 Standardize Department names.
df['Department'] = df['Department'].str.strip();
df.loc[(df['Department']=="it") | (df['Department']=="I.T"), 'Department'] = "IT";
df.loc[df['Department']=="sales", 'Department'] = "Sales";

print(df['Department']);

#Q8 Standardize City names.
df['City'].unique()
df['City'] = df['City'].str.strip();
df['City'] = df['City'].str.capitalize();

#Q9 Identify age outliers.
q1 = df['Age'].quantile(0.25);
q3 = df['Age'].quantile(0.75);
print(q1,q3);

ll = (q1 - 1.5*(q3-q1));
ul = (q3+(1.5*(q3-q1)));

print(df.loc[(df['Age']<ll) | (df['Age']>ul), 'Age']);

# Treatment of Age Outliers
df.loc[df['Age']<ll, 'Age'] = ll;
df.loc[df['Age']>ul, 'Age'] = ul;

#Q10 Identify salary outliers.

q1 = df['Salary'].quantile(0.25);
q3 = df['Salary'].quantile(0.75);
iqr = (q3-q1);

print(q1, q3, iqr);

lower_limit = (q1 - (1.5*iqr));
upper_limit = (q3 + (1.5*iqr));

df.loc[(df['Salary']<lower_limit) | (df['Salary']> upper_limit)];

# Treatment of Salary Outliers
df.loc[df['Salary']<lower_limit, 'Salary'] = lower_limit;
df.loc[df['Salary']>upper_limit ,'Salary'] = upper_limit;

#Q11 Display unique departments.
print(df['Department'].unique());
#Q12 Display unique cities.
print(df['City'].unique());
#13  Save the cleaned dataset to a new Excel file.
df.to_excel("cleaned_data.xlsx", index=False);

```
