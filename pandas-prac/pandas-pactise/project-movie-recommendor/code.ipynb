```python
import pandas as pd
import numpy as np
```

```python
# Q1 Read movie dataset.
df = pd.read_excel("movie_recommender_50_movies.xlsx");

# Q2 Display shape.
print(df.shape);
# Q3 Display column names.
print(df.columns.to_list());
# Q4 Check missing values.
print(df.isnull().sum());
# Q5 Remove duplicate movies.
df.drop_duplicates(inplace=True);
```

```python
# Q6 Display all Sci-Fi movies.
print(df.loc[df['Genre']=='Sci-Fi']);
# Q7 Display all Action movies.
print(df.loc[df['Genre']=='Action']);
# Q8 Display movies with rating > 8.
print(df.loc[df['Rating']>8]);
# Q9 Display movies released after 2010.
print(df.loc[df['Year']>2010]);
# Q10 Display Romance movies with rating > 7.
print(df.loc[(df['Genre']=='Romance') & (df['Rating']>7)]);
```

```python
# Q11 Sort movies by Rating descending.
print(df.sort_values(by='Rating', ascending=False));
# Q12 Display Top 5 movies.
print(df.sort_values(by='Rating', ascending=False).head(5));
# Q13 Display Lowest Rated movie.
print(df.sort_values(by='Rating').head(1));
```

```python
# Q14 Count movies genre-wise.

print(df.groupby('Genre')['MovieName'].count());

# Q15 Find average rating genre-wise.
print(df.groupby('Genre')['Rating'].mean());

# Q16 Find highest-rated movie in each genre.

genre = df.groupby('Genre')['Rating'].max().index.to_list();
i = 0;

for item in df.groupby('Genre')['Rating'].max():
    # print(item, genre[i]);
    
    print(df.loc[(df['Genre']==genre[i]) & (df['Rating'] == item) ]);
    i+=1;
```

```python
# Q17 Display all Sci-Fi movies.
print(df.loc[df['Genre']=='Sci-Fi']);

# Q18 Sort by rating descending.
print(df.loc[df['Genre']=='Sci-Fi'].sort_values(by='Rating', ascending=False));

# Q19 Recommend top 3 Sci-Fi movies.
print(df.loc[df['Genre']=='Sci-Fi'].sort_values(by='Rating', ascending=False).head(3));

```

```python
# Q20 Find genre having maximum movies.
print(df['Genre'].value_counts().sort_values(ascending=False).index[0]);
```

```python
# Final Mini Project

genre_input = 'Action';

index = df.loc[df['Genre']==genre_input].sort_values(by='Rating', ascending=False).head(3).index.to_list();
print("Below are 3 recommended movies");

for movie in df.loc[index, 'MovieName']:
    print(movie);
```
