```python
import pandas as pd
import numpy as np
```

```python
# Q1 Read Employees and NewEmployees sheets.

df_emp = pd.read_excel("pandas_merge_concat_practice.xlsx", sheet_name="Employees");
df_Nemp = pd.read_excel("pandas_merge_concat_practice.xlsx", sheet_name="NewEmployees");
print(df_emp);
print(df_Nemp);

# Q2 Concatenate both DataFrames vertically.

df_concat = pd.concat([df_emp, df_Nemp], axis=0);

# Q3 Display total number of employees after concatenation.
print(df_concat);
print(len(df_concat));

# Q4 Display only Name and Salary columns after concatenation.

print(df_concat[['Name', 'Salary']]);

# Q5 Save the concatenated DataFrame to all_employees.xlsx.
df_concat.to_excel("all_employees.xlsx", index=False)
```

```python
# Q6 Read Employees and Departments sheets.

df_emp = pd.read_excel("pandas_merge_concat_practice.xlsx", sheet_name="Employees");
df_dept = pd.read_excel("pandas_merge_concat_practice.xlsx", sheet_name="Departments");

print(df_emp);
print(df_dept);

# Q7 Merge both DataFrames using DeptID.

df_merge = pd.merge(df_emp,df_dept, on="DeptID");

# Q8 Display only Name and Department columns after merge.
print(df_merge[['Name', 'Department']]);

# Q9 Display employees belonging to IT department after merge.
print(df_merge.loc[df_merge['Department']=='IT']);

# Q10 Display Name, Salary and Department after merge.
df_merge[['Name', 'Salary', 'Department']]
```
