```python
import pandas as pd
```

```python
df = pd.read_excel("pandas_assignment_dataset.xlsx");

df
```

```python
#Q1
df.loc[0,'Marks'] = 100

#Q2
df.loc[df['Name'] == 'Rama', 'Education'] = 'M.Tech'


#Q3
df.iloc[-1, -1] = 'Pune'

#Q4
df.iloc[1, -2] = 999

#Q5
df.iloc[2, 0] = 'Raj'

#Q6
df['Marks'] = df['Marks'] + 10

#Q7
df['Marks'] = df['Marks'] * 2


df
```

```python
#Q8
df['DoubleMarks'] = df['Marks'] * 2

#Q9
df['Name'] = df['Name'].str.upper()

#Q10
df['Education'] = df['Education'].str.lower()

#Q11

df['Country'] = 'India';

#Q12

df['Tax'] = df['Marks'] * 0.18;

#Q13
df['Marks'] = df['Marks'] + 5;

#Q14
df['Rank'] = list(range(1,11));

#Q15

df.loc[df['Education']=="mba", 'Education'] = 'MASTER';

#Q16
df['Grade'] = ['A','B', 'C','D','E', 'F', 'G','H', 'I', 'J']

#Q17
city = df['City'].tolist();

df['City'] = city;

#Q18

remarks = ["Good", "Bad", "Average", "Very Good", "Excellent"];
new = [];
from random import choice
import numpy as np
for i in range(1,11):
    new.append(choice(remarks));

df['Remarks'] = new;

#Q19
numbers = [1,2,3,4,5,6,7];
lucky_numbers = [];
for i in range(1,11):
    lucky_numbers.append(choice(numbers));

df['LuckyNumber'] = lucky_numbers;

#Q20
res = ["Pass", "Fail"];
results = [];
for i in range(1,11):
    results.append(choice(res));

df['Result'] = results;

#Q21
df.loc[df['Marks']<50, 'Result'] = 'Fail'

#Q22
df.loc[df['Marks']>=50, 'Result'] = 'Pass'

#Q23
df.loc[df['Marks']>95, 'Result'] = 'Topper'

#Q24
df.loc[(df['Marks']>90) & (df['Marks']<100) , 'Result'] = 'Excellent';

#Q25
df.loc[(df['Marks']>90)&(df['Education']=='b.tech'), 'Education'] = 'BTECH';

df
```
