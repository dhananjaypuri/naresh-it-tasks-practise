```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("pandas_data_cleaning_practice.xlsx");

# Q1 Display DataFrame shape.
print(df.shape)

# Q2 Display column names.
print(df.columns.to_list());

# Q3 Display data types.
print(df.dtypes)

# Q4 Find missing values column-wise.

print(df.isna().sum(axis=0));
# Q5 Display rows containing missing values.
print(df.loc[df.isna().any(axis=1)]);

# Q6 Replace NaN Age with 18.
df.loc[df['Age'].isna(), 'Age'] = 18;

# Q7 Replace NaN Salary with 0.
df.loc[df['Salary'].isna(), 'Salary'] = 0;
# Q8 Find duplicate rows.
print(df.loc[df.duplicated()]);
# Q9 Count duplicate rows.
print(len(df.loc[df.duplicated()]));
# Q10 Remove duplicate rows.

df.drop_duplicates(inplace=True);

# # Q11 Remove leading/trailing spaces from Department.
df['Department'] = df['Department'].str.strip();
# # Q12 Convert Department to uppercase.
df['Department'] = df['Department'].str.upper();
# # Q13 Display unique departments.
print(df['Department'].unique());
# # Q14 Count employees department-wise.
print(df.groupby('Department')['Name'].count());
# Q15 Save the cleaned data to: cleaned_employee_data.xlsx
df.to_excel("cleaned_employee_data.xlsx", index=False);
```
