```python
import pandas as pd
```

```python
df = pd.DataFrame({'city':["Hyd","Hyd","Hyd","Hyd","Hyd","Chennai"]})
df
print(df['city'].unique().tolist());
print(df['city'].value_counts());

print("Number of Values in column" , len(df['city']));

df['city'].value_counts()/len(df['city'])

mapping = {"Hyd": (df['city'].value_counts()/len(df['city'])).loc['Hyd'],
           "Chennai": (df['city'].value_counts()/len(df['city'])).loc['Chennai']};

df['city'] = df['city'].map(mapping);
df
```

```python
df = pd.DataFrame({
    'city': [
        'Hyderabad',
        'Mumbai',
        'Hyderabad',
        'Delhi',
        'Mumbai',
        'Hyderabad',
        'Pune',
        'Delhi',
        'Mumbai',
        'Hyderabad'
    ]
})
df['city'].unique().tolist();
df['city'] = df['city'].map(df['city'].value_counts()/len(df['city']));
df
```
