```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({'shirt_size':["S","M","S","XL","S","L"]})

# categories = df['shirt_size'].unique().tolist();

oe = OrdinalEncoder(categories=[['S', 'M', 'L', 'XL']]);
df['shirt_size_en'] = oe.fit_transform(df[['shirt_size']]);

df
```

```python
df = pd.DataFrame({
    'education': [
        'Graduate',
        'PhD',
        'School',
        'Masters',
        'Graduate'
    ]
});

df['education'].unique().tolist()

categories = ['School', 'Graduate', 'Masters', 'PhD'];

oe = OrdinalEncoder(categories=[categories], dtype=int);
df['education_enc'] = oe.fit_transform(df[['education']]);
df.sort_values(by='education_enc', inplace=True)
df
```

```python
df = pd.DataFrame({
    'priority': [
        'High',
        'Critical',
        'Low',
        'Medium',
        'High'
    ]
});

categories = ['Low', 'Medium', 'High', 'Critical'];
oe = OrdinalEncoder(categories=[categories]);

df['priority_en'] = oe.fit_transform(df[['priority']]);
df.sort_values(by='priority_en', inplace=True);

df
```
