```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("employee_valuecounts_sorting_dataset.xlsx");
# print(df)

# type(df.groupby('Department')['Name'].unique())
# Q1 Display the number of employees in each department.
print(df.groupby('Department')['Name'].count());

# Q2 Display the maximum salary in each department.
print(df.groupby('Department')['Salary'].max());

# Q3 Display the minimum salary in each department.
print(df.groupby('Department')['Salary'].min());

# Q4 Display the average salary in each department.
print(df.groupby('Department')['Salary'].mean());

# Q5 Display the total salary paid in each department.
print(df.groupby('Department')['Salary'].sum());

# Q6 Display the employee count in each department.
print(df.groupby('Department')['Name'].count());
```

```python
# Q7 Find the department having the highest average salary.
print(df.groupby('Department')['Salary'].mean().sort_values(ascending=False).index[0]);

# Q8 Display maximum and minimum salary department-wise in a single query.
print(df.groupby('Department')['Salary'].agg(['max', 'min']));
```
