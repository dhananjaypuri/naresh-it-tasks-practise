```python
import pandas as pd
import numpy as np
```

```python
# Reading Excel
# Q1
# Read employee.xlsx.

df = pd.read_excel("pandas_exercise_employee.xlsx");

# Q2
# Display complete DataFrame.
print(df);

# Q3
# Display first 2 rows.
print(df.head(2));

# Q4
# Display last 3 rows.
print(df.tail(3));

# Q5
# Print column names.
cl = df.columns.to_list();
print(cl);

# Q6
# Print shape of DataFrame.
print(df.shape);
print("Rows:", df.shape[0]);
print("Columns", df.shape[1]);

# Q7
# Print data types.
print(df.dtypes);

# Q8
# Display only first row.
print(df.iloc[0]);

# Q9
# Display only last row.
print(df.iloc[-1]);

# Q10
# Display rows from index 1 to 3.

print(df.loc[1:3]);
```

```python
# Extract Columns
# Q11
# Display Name column.
print(df['Name']);

# Q12
# Display Salary column.
print(df['Salary']);

# Q13
# Display Name and Department columns.
print(df[['Name', 'Department']]);

# Q14
# Display Age and Salary columns.
print(df[['Age', 'Salary']]);

# Q15
# Display EmpID, Name and Salary columns.
print(df[['EmpID', 'Name', 'Salary']]);
```

```python
# Filtering Rows
# Q16
# Display employees from IT department.
print(df[df['Department']=='IT']);

# Q17
# Display employees from HR department.
print(df[df['Department']=='HR']);

# Q18
# Display employees whose age is greater than 25.
print(df[df['Age']>25]);

# Q19
# Display employees whose salary is greater than 55000.
print(df[df['Salary']>55000]);

# Q20
# Display employees whose department is Finance.
print(df[df['Department']=='Finance']);

# Q21
# Display employees whose age is less than 30.
print(df[df['Age']<30]);

# Q22
# Display employees whose name is Emma.
print(df[df['Name']=='Emma']);

# Q23
# Display employees whose EmpID is 103.
print(df[df['EmpID']==103]);

# Q24
# Display employees whose salary is not null.
print(df[df['Salary'].notna()]);

# Q25
# Display employees whose age is null.
print(df[df['Age'].isna()]);
```

```python
# Add New Columns
# Q26
# Add Bonus column with value 5000.
df['Bonus'] = 5000;
# Q27
# Add Status column with value Active.
df['Status'] = 'Active';
# Q28
# Add Country column with value India.
df['Country'] = 'India';
# Q29
# Add Experience column with value 0.
df['Experience'] = 0;

# Q30
# Add Grade column with value A.
df['Grade'] = 'A';
df
```

```python
# Update Existing Values
# Q31
# Update John's salary to 65000.
df.loc[df['Name']=='John', 'Salary'] = 65000;

# Q32
# Update Emma's department to Finance.
df.loc[df['Name']=='Emma', 'Department'] = 'Finance';

# Q33
# Update Bob's age to 29.
df.loc[df['Name']=='Bob', 'Age'] = 29;

# Q34
# Update EmpID 104 salary to 90000.
df.loc[df['EmpID']==104, 'Salary'] = 90000;

# Q35
# Change all HR values to Human Resources.
df.loc[df['Department']=='HR', 'Department'] = 'Human Resources';

# Q36
# Change all IT values to Technology.
df.loc[df['Department']=='IT', 'Department'] = 'Technology';

# Q37
# Update Status column to Inactive.
df['Status'] = "Inactive";

# Q38
# Replace all NaN salaries with 0.
df.loc[df['Salary'].isna(), 'Salary'] = 0;

# Q39
# Replace all NaN ages with 18.
df.loc[df['Age'].isna(), 'Age'] = 18;

# Q40
# Change Country value from India to USA.
df['Country'] = 'USA';
```

```python
# loc Practice
# Q41
# Update salary of row index 2 to 70000 using loc.
df.loc[2,'Salary'] = 70000;

# Q42
# Update department of row index 1 to Sales.
df.loc[1, 'Department'] = 'Sales';

# Q43
# Update age where Name == "David".
df.loc[df['Name']=='David', 'Age'] = 29;

# Q44
# Update salary where EmpID == 105.
df.loc[df['EmpID']==105, 'Salary'] = 60000;

# Q45
# Update Grade where Department == "Finance".
df.loc[df['Department']=='Finance', 'Grade'] = 'B';
```

```python
# Rename Operations
# Q46
# Rename Department to Dept.
df.rename(columns={'Department': 'Dept'});

# Q47
# Rename Salary to MonthlySalary.
df.rename(columns={'Salary': 'MonthlySalary'});

# Q48
# Rename Age to EmployeeAge.
df.rename(columns={'Age': 'EmployeeAge'});

# Q49
# Rename Name to EmployeeName.
df.rename(columns={'Name': 'EmployeeName'});

# Q50
# Rename EmpID to EmployeeID.
df.rename(columns={'EmpID': 'EmployeeID'});
```

```python
# Drop Operations
# Q51
# Drop Bonus column.
df.drop(columns='Bonus');

# Q52
# Drop Status column.
df.drop(columns='Status')

# Q53
# Drop Age column.
df.drop(columns='Age');

# Q54
# Drop rows where Name is Emma.
df.drop(index=df.loc[df['Name']=='Emma'].index.to_list());

# Q55
# Drop rows where Department is HR.
df.drop(index=df.loc[df['Department']=='HR'].index);

# Q56
# Drop first row.
df.drop(index=0);

# Q57
# Drop last row.
df.drop(index=len(df)-1);

# Q58
# Drop rows with null values.
df.drop(index=df[df.isna().any(axis=1)].index);

# Q59
# Drop EmpID and Salary columns.
df.drop(columns=['EmpID', 'Salary']);

# Q60
# Drop all rows where Salary is 0.
df.drop(index=df[df['Salary']==0].index);
```

```python
# Save Excel
# Q61
# Save DataFrame to output.xlsx.
df.to_excel("output.xlsx");

# Q62
# Save without index.
df.to_excel("output.xlsx", index=False);

# Q63
# Save only Name and Salary columns.
df2 = df[['Name', 'Salary']];
df2.to_excel("new.xlsx", index=False);

# Q65
# Save only Finance employees.
df3 = df.loc[df['Department']=='Finance'];

df3.to_excel("new.xlsx", index=False);
```
