```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder , OneHotEncoder , LabelEncoder
```

```python
# Q1 Load LoanData.csv into a DataFrame.
df = pd.read_csv("LoanData.csv");
```

```python
# Q2 Print:

# Shape
# Columns
# Data types
# Head
# Tail

print(df.shape);
print(df.columns.to_list());
print(df.dtypes);
print(df.head());
print(df.tail());
```

```python
# Q3 Generate a dataset summary using:

print(df.info());
print(df.describe());
```

```python
# Q4 Find missing values in every column.
print(df.isna().sum());
# Q5 Calculate missing value percentage.
print(df.isna().sum()*100/len(df));
# Q6 Identify duplicate records.
print(df.duplicated().sum());
#Q7Check whether Loan_ID contains duplicates.
print(df['Loan_ID'].duplicated().sum());

```

```python
# Q8Handle missing values:
# Numerical → Median
# Categorical → Mode
columns = df.columns.to_list();
## Neueric Columns
for col in columns:
    if df[col].dtypes in ['int64', 'float64']:
        # print("Neumeric", col);
        df[col] = df[col].fillna(df[col].median())

## How to find categorical columns

for catcol in columns:
    if df[catcol].dtypes in ['str']:
        # df[col] = df[col].fillna(df[col].median())
        if len(df[catcol].unique())<=10:
            df[catcol] = df[catcol].fillna(df[catcol].mode()[0]);

```

```python
# Q9 Remove duplicate records.
df.drop_duplicates(inplace=True);
df
```

```python
# Q10 Verify dataset contains:# No missing values # No duplicates
print(df.isna().sum());
print(df[df.duplicated()]);
```

```python
# Q11 Separate columns into: Numerical Categorical
neumeric_cols = [];
cat_cols = [];
for col in columns:
    if df[col].dtypes in ['int64', 'float64']:
        neumeric_cols.append(col);
    elif df[col].dtypes in ['str']:
        if len(df[col].unique())<=10:
            cat_cols.append(col);

print(neumeric_cols);
print(cat_cols);

# Q12 Find unique values of every categorical column.
for col in cat_cols:
    print(df[col].unique());

```

```python
# Q14 Detect outliers in:
# ApplicantIncome
# CoapplicantIncome
# LoanAmount using IQR.
outliers_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount'];
def find_outliers(col):
    q1 = df[col].quantile(0.25);
    q3 = df[col].quantile(0.75);
    IQR = q3 - q1;
    ll = (q1-(1.5*IQR));
    ul = (q3+(1.5*IQR));

    # print(ll);
    # print(ul);

    outliers = df.loc[(df[col]<ll) | (df[col]>ul), col];
    return {'outliers': outliers, 'll': ll, 'ul': ul}

for c in outliers_cols:
    result = find_outliers(c);
    # Q15 Count outliers.
    print(f"Outliers in {c}", len(result['outliers']));
    # Q16. Treat Outliers
    df[c] = df[c].astype('float');
    df.loc[(df[c]<result['ll']), c] = result['ll'];
    df.loc[(df[c]>result['ul']), c] = result['ul'];




```

```python
# Q18

# Create TotalIncome

# TotalIncome = ApplicantIncome + CoapplicantIncome

df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome'];
# Q19
# Create IncomePerDependent
# TotalIncome / Dependents
df['Dependents'] = df['Dependents'].replace({'3+': 3}).astype('int');

arr = [];
print(len(df['Dependents']))
print(len(df['TotalIncome']))
i = 0;
for inc in df['TotalIncome']:
    item = 0;
    if df.iloc[i, 3] == 0:
        item = df.iloc[i, 13];
    else:
        item = df.iloc[i, 13]/df.iloc[i, 3];
    i += 1;

    arr.append(item);
df['IncomePerDependent'] = np.array(arr);
df['IncomePerDependent']
```

```python
# Q21 # Create LoanIncomeRatio
# LoanAmount / TotalIncome
df['LoanAmount'] = df['LoanAmount'] * 1000;
df[['LoanAmount', 'TotalIncome']].head()

df['LoanIncomeRatio'] = df['LoanAmount'] / df['TotalIncome'];


# Q22
# Create Income Category:
# avg = df['TotalIncome'].mean();
# Low
# Medium
# High

avg = df['TotalIncome'].mean()
print(avg)
df.loc[(df['TotalIncome'] < avg), 'Inc_Category'] = "low";
df.loc[((df['TotalIncome'] >avg) & (df['TotalIncome'] < avg +1000)), 'Inc_Category'] = "medium";
df.loc[(df['TotalIncome'] > avg+1000), 'Inc_Category'] = "high";

df.head()

df['Inc_Category'].value_counts()

# Q23 # Create Loan Term Years # Loan_Amount_Term / 12

df['Loan_Amount_Term'] = df['Loan_Amount_Term']/12;



```

```python
#Q25. Encode Gender

df['Gender'].unique()
le = LabelEncoder();

df['Gender'] = le.fit_transform(df['Gender']);

# Q26. Encode Married
df['Married'] = df['Married'].replace({'Yes': 1, 'No': 0});

# Q27. Encode Self_Employed
df['Self_Employed'].unique()
df['Self_Employed'] = le.fit_transform(df['Self_Employed']);

# Q28. Encode Education
edu_catergories = df['Education'].unique().tolist();
oe = OrdinalEncoder(categories=[['Not Graduate', 'Graduate']], dtype=int)
df['Education'] = oe.fit_transform(df[['Education']]);

# Q29. Encode Loan_Status (Target Variable)
df['Loan_Status'].unique()
df['Loan_Status'] = le.fit_transform(df['Loan_Status']);
df['Loan_Status'].unique()

# Q30. Encode Property_Area

df = pd.get_dummies(df,columns=['Property_Area'], dtype=int);

```
