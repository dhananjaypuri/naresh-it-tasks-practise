```python
import pandas as pd
import numpy as np
```

```python
# Q1 Perform an inner join between Employees and Departments.

df_emp = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Employees");
df_dept = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Departments");

print(df_emp);
print(df_dept);

df_inner = pd.merge(df_emp, df_dept, on="DeptID", how="inner");
print(df_inner);

# Q2 How many rows are returned by the inner join?
print(len(df_inner));

# Q3 Which employee is removed during the inner join?

print("David is removed");
```

```python
# Q4 Perform a left join between Employees and Departments.
df_emp = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Employees");
df_dept = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Departments");

print(df_emp);
print(df_dept);

df_left = pd.merge(df_emp, df_dept, on="DeptID", how="left");
print(df_left);
# Q5 Which employee has a null department after the join?
print(df_left[df_left.isna().any(axis=1)]);
# Q6 How many rows are returned by the left join?
print(len(df_left));
```

```python
# Q7 Perform a right join between Employees and Departments.
df_emp = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Employees");
df_dept = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Departments");
print(df_emp);
print(df_dept);
df_right = pd.merge(df_emp,df_dept, on="DeptID", how="right");
print(df_right);
# Q8 Which departments have no employees?
print(df_right.loc[df_right['Name'].isna(), 'Department']);
# Q9 How many rows are returned by the right join?
print(len(df_right));
```

```python
# Q10 Perform a full outer join.
df_emp = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Employees");
df_dept = pd.read_excel("pandas_joins_practice.xlsx", sheet_name="Departments");
print(df_emp);
print(df_dept);
df_outer = pd.merge(df_emp, df_dept, on="DeptID", how="outer");
print(df_outer);
# Q11 How many rows are returned?
print(len(df_outer));
# Q12 Identify: employees without departments , departments without employees
print(df_outer.loc[df_outer['Department'].isna(), 'Name']);
print(df_outer.loc[df_outer['Name'].isna(), 'Department']);
```
